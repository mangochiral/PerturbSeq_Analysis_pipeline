```python
import scanpy as sc
import os, sys
import numpy as np
import pandas as pd
import glob
```

```python
dir_list = pd.read_csv('/Users/chandrima.modak/Gladstone Dropbox/Chandrima Modak/gw-CRISPRa_from_cluster/h5ad_raw_files_info.csv')
lane_cols_for_eff = [c for c in dir_list.columns if c.startswith("lane")]
run = []
for _, row in dir_list[lane_cols_for_eff].iterrows():
    for col in lane_cols_for_eff:
        v = row[col]
        run.append(f"{v}_{col}")
```

```python
path = '/Users/chandrima.modak/Gladstone Dropbox/Chandrima Modak/gw-CRISPRa_from_cluster'
for i in run:
    name = i.split('_')[2]
    pattern = glob.glob(os.path.join(path, i, '*_crispr_preprocessed.h5ad'))[0]
    crispr_a = sc.read_h5ad(pattern)
    if 'i' in name:
        crispr_a.var['perturbation_type'] = np.where(crispr_a.var_names.str.startswith('NTC'), 'NTC', 'CRISPRi')
        crispr_a = crispr_a[:, crispr_a.var['perturbation_type'] != 'CRISPRa'].copy()
        crispr_a.write_h5ad(os.path.join(path, f'{i}_crispr_preprocessed.h5ad')
    elif 'a' in name:
        crispr_a.var['perturbation_type'] = np.where(crispr_a.var_names.str.startswith('NTC'), 'NTC', 'CRISPRa')
        crispr_a = crispr_a[:, crispr_a.var['perturbation_type'] != 'CRISPRi'].copy()
        crispr_a.write_h5ad(os.path.join(path, f'{i}_crispr_preprocessed.h5ad')
```
